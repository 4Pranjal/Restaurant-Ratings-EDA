```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how ='left')
```

```python
final_df.head()
```

```python
# to check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
# to check the records number
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# to create a pie chart
plt.pie(country_val,labels=country_names)
```

```python
# to create a pie chart - top 3 country
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
#observvation
'''
When rating is between 4.5 to 4.9 is excellent
3.9 to 4.5 =>very good
3.5 to 3.9 =>good
'''
```

```python
ratings.head()
```

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','lightgreen','green'])
```

```python
# observation
'''
1. not rated counting is very high
2. maximum number of rating is from 2.5 to 3.4
'''
```

```python
# count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','lightgreen','green'])
```
